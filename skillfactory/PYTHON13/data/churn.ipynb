```python
import pandas as pd 
import matplotlib.pyplot as plt
churn_data = pd.read_csv('C:\IDE\skillfactory\PYTHON13\data\churn.csv')
churn_data.head()
```

```python
# подсчет количества ушедших и лояльных клиентов
exited = churn_data[churn_data['Exited'] == 1]['Exited'].count()
loyal = churn_data[churn_data['Exited'] == 0]['Exited'].count()

# построение графика
plt.pie([exited, loyal], labels=['Ушедшие', 'Лояльные'], autopct='%1.1f%%')
plt.title('Соотношение ушедших и лояльных клиентов')
plt.show()
```

```python
# фильтрация данных по условию баланса больше 2 500 долларов
filtered_data = churn_data[churn_data['Balance'] > 2500]

# построение гистограммы распределения баланса
plt.hist(filtered_data['Balance'], bins=20)
plt.xlabel('Баланс')
plt.ylabel('Частота')
plt.title('Распределение баланса пользователей')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# загрузка данных из файла
churn_data = pd.read_csv('C:\IDE\skillfactory\PYTHON13\data\churn.csv')

# выбор данных для лояльных клиентов
loyal_data = churn_data[churn_data['Exited'] == 0]

# выбор данных для ушедших клиентов
churned_data = churn_data[churn_data['Exited'] == 1]

# построение графиков распределения баланса для каждой группы клиентов
sns.histplot(data=loyal_data, x='Balance', kde=True)
sns.histplot(data=churned_data, x='Balance', kde=True)
```

```python
# построение коробчатой диаграммы для возраста в каждой группе клиентов
sns.boxplot(data=churn_data, x='Exited', y='Age')
plt.show()
```

```python
# построение scatter plot для взаимосвязи кредитного рейтинга и заработной платы с расцветкой по признаку оттока клиентов
sns.scatterplot(data=churn_data, x='CreditScore', y='EstimatedSalary', hue='Exited')
plt.show()

```

```python
# построение bar plot для сравнения оттока клиентов по полу
sns.barplot(data=churn_data, x='Gender', y='Exited')
plt.show()
```

```python
# построение многоуровневой столбчатой диаграммы для сравнения оттока клиентов по числу приобретенных услуг
sns.barplot(data=churn_data, x='NumOfProducts', y='Exited')
plt.show()
```

```python
# построение столбчатой диаграммы для сравнения оттока клиентов в зависимости от наличия статуса активного клиента
sns.barplot(data=churn_data, x='IsActiveMember', y='Exited')
plt.show()
```
